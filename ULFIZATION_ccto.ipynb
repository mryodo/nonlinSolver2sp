```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
d12=np.linspace(0, 0.001, 100)
sm2=np.linspace(0, 0.2, 100)
N1=np.loadtxt('N1cctoD1.txt', delimiter=',')
N2=np.loadtxt('N2cctoD1.txt', delimiter=',')
for i in range(N1.shape[0]):
    for j in range(N1.shape[1]):
        if math.isnan(N1[i][j]):
            N1[i][j]=N1[i][j+1]
for i in range(N2.shape[0]):
    for j in range(N2.shape[1]):
        if math.isnan(N2[i][j]):
            N2[i][j]=N2[i][j+1]
            
N1=N1.T
N2=N2.T

N1=np.flipud(N1)
N2=np.flipud(N2)
```

```python
X, Y = np.meshgrid(d12, sm2)


cmap = plt.cm.get_cmap("winter")
cmap.set_under("magenta")
cmap.set_over("yellow")

plt.figure(figsize=(10, 10))

plt.contourf(X, Y, np.exp(N1/10), 1000, cmap='Blues', alpha=1, origin='lower', extend='both')
plt.colorbar()
plt.locator_params(nbins=4)

plt.figure(figsize=(10, 10))
plt.contourf(X, Y, np.log(N2+1), 1000, cmap='Reds', alpha=1, origin='lower', extend='both')
plt.colorbar()
plt.locator_params(nbins=4)

plt.show()
```
