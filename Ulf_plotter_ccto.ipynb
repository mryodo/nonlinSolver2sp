```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import math
```

```python
from matplotlib.colors import Normalize

class MidpointNormalize(Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

```python
d12=np.linspace(0, 0.001, 100)
sm2=np.linspace(0, 0.2, 100)
N1=np.loadtxt('N1cctoD1.txt', delimiter=',')
N2=np.loadtxt('N2cctoD1.txt', delimiter=',')

for i in range(N1.shape[0]):
    for j in range(N1.shape[1]):
        if math.isnan(N1[i][j]):
            N1[i][j]=N1[i][j+1]
        if N1[i][j]<0:
            N1[i][j]=1e-5

for i in range(N2.shape[0]):
    for j in range(N2.shape[1]):
        if math.isnan(N2[i][j]):
            N2[i][j]=N2[i][j+1]
        if N2[i][j]<0:
            N2[i][j]=1e-5
N1=N1.T
N2=N2.T
N1=np.flipud(N1)
N2=np.flipud(N2)   
            
norm = MidpointNormalize(midpoint=250)
norm2 = MidpointNormalize(midpoint=0)

d12, sm2=np.meshgrid(d12, sm2)

plt.figure(figsize=(15, 10))
cmap = plt.cm.get_cmap("Blues")
plt.contourf(d12,sm2,N1,1000, norm=norm, cmap=cmap, alpha=0.9, antialiased=False)
plt.colorbar()
cmap = plt.cm.get_cmap("Reds")
plt.contourf(d12,sm2,N2,1000, norm=norm, cmap=cmap
            , alpha=0.2, antialiased=False)
plt.colorbar()


plt.show()
```
